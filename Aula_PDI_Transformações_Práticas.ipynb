#Prática 3: Gerando uma imagem negativa

```python
image_pratica_3 = io.imread('pratica3.jpeg', as_gray=True)
h,w = image_pratica_3.shape
print(f'altura:{h}, largura:{w}')
```

```python
new_image = image_pratica_3.copy()

for i in range(h):
  for j in range(w):
    new_image[i,j] = 255 - image_pratica_3[i,j]

plt.imshow(new_image, cmap='gray')
```

# Prática 4: Remoção de Ruída com Filtro da Média e da Mediana

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
```

```python
image = io.imread('pratica4.png', as_gray=True)
h,w = image.shape
print(f'altura:{h}, largura:{w}')
```

```python
filtered_image = image.copy()
```

```python
for i in range(1, h):
  for j in range(1, w):
    filtered_image[i,j] = np.median(image[i-1:i+2, j-1:j+2])
```

```python
plt.imshow(filtered_image, cmap='gray')
```

# Prática 5: Gerando Ruído Salt and Pepper na Imagem

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
import random
```

```python
image = io.imread('pratica3.jpeg', as_gray=True)
h,w = image.shape
print(f'altura:{h}, largura:{w}')
```

```python
def salt_and_pepper(image, num_salt, num_pepper):
  output = image.copy()
  for _ in range(num_salt):
    i = random.randint(0, image.shape[0]-1)
    j = random.randint(0, image.shape[1]-1)
    output[i,j] = 255
  for _ in range(num_pepper):
    i = random.randint(0, image.shape[0]-1)
    j = random.randint(0, image.shape[1]-1)
    output[i,j] = 0
  return output

num_salt = 500
num_pepper = 500

noisy_image = salt_and_pepper(image, num_salt, num_pepper)
io.imshow(noisy_image)
plt.show()
```
